```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ASUS\Documents\-em-bond-macro-analysis\data\Cleaned\panel_monthly_2010_2024_copy.csv")
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['country','date'])
numeric_cols = ['yield_10y','CPI_annual_pct','GDP_growth_annual_pct','External_debt_usd','usd_index','brent']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```

```python
df
```

```python
# To see how various variables move with time. Are there any trends, spikes or seasonality? Check for distributions and outliers. 
```

```python
# Date vs Yield

import matplotlib.pyplot as plt
import seaborn as sns

countries = df['country'].unique()

for c in countries:
    temp = df[df['country']==c]
    plt.figure(figsize=(12,6))
    plt.plot(temp['date'], temp['yield_10y'], marker='o')
    plt.title(f'10-Year Yield Over Time - {c}')
    plt.xlabel('Date')
    plt.ylabel('10-Year Yield (%)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# Date vs CPI

countries = df['country'].unique()

for c in countries:
    temp = df[df['country']==c]
    plt.figure(figsize=(12,6))
    plt.plot(temp['date'], temp['CPI_annual_pct'], marker='o')
    plt.title(f'CPI Over Time - {c}')
    plt.xlabel('Date')
    plt.ylabel('CPI Annual (%)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# CPI, External Debt and GDP Growth have been forward filled for all the months of the year because of the annual nature of their values
```

```python
# Date vs GDP Growth

countries = df['country'].unique()

for c in countries:
    temp = df[df['country']==c]
    plt.figure(figsize=(12,6))
    plt.plot(temp['date'], temp['GDP_growth_annual_pct'], marker='o')
    plt.title(f'GDP Growth Over Time - {c}')
    plt.xlabel('Date')
    plt.ylabel('GDP Growth Annual (%)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# Date vs External Debt

countries = df['country'].unique()

for c in countries:
    temp = df[df['country']==c]
    plt.figure(figsize=(12,6))
    plt.plot(temp['date'], temp['External_debt_usd'], marker='o')
    plt.title(f'External Debt Over Time - {c}')
    plt.xlabel('Date')
    plt.ylabel('External Debt USD')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# Debt/GDP ratio gives a better picture of whether a country will be able to pay for the bonds issued. But unavailability of nominal GDP values led to individual understanding of GDP Growth and External Debt.
```

```python
# Date vs Brent Crude Oil Prices

countries = df['country'].unique()

for c in countries:
    temp = df[df['country']==c]
    plt.figure(figsize=(12,6))
    plt.plot(temp['date'], temp['brent'], marker='o')
    plt.title(f'Brent Crude Oil Prices Over Time - {c}')
    plt.xlabel('Date')
    plt.ylabel('Brent Crude Oil Prices')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# Correlation Matrix for a quick summary of linear relationships
```

```python
for c in countries:
    temp = df[df['country']==c]
    corr = temp[numeric_cols].corr()
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Matrix - {c}')
    plt.show()
```
